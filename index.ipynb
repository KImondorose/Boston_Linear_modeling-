### Boston Housing Dataset Column Descriptions

1. **CRIM**: Per capita crime rate by town.
2. **ZN**: Proportion of residential land zoned for lots over 25,000 sq. ft.
3. **INDUS**: Proportion of non-retail business acres per town.
4. **CHAS**: Charles River dummy variable (1 if tract bounds river; 0 otherwise).
5. **NOX**: Nitric oxides concentration (parts per 10 million).
6. **RM**: Average number of rooms per dwelling.
7. **AGE**: Proportion of owner-occupied units built before 1940.
8. **DIS**: Weighted distances to five Boston employment centers.
9. **RAD**: Index of accessibility to radial highways.
10. **TAX**: Full-value property tax rate per $10,000.
11. **PTRATIO**: Pupil-teacher ratio by town.
12. **B**: 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.
13. **LSTAT**: Percentage lower status of the population.
14. **MEDV**: Median value of owner-occupied homes in $1000s (target variable).

## Get the Outliers 

```python
# Calculate IQR for selected columns
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = ((df < lower_bound) | (df > upper_bound)).any(axis=1)

# Print indices of outliers
outlier_indices = df.index[outliers]
```

## Remove the outliers 

```python
# droping all columns with outliers 
df = df.drop(outlier_indices)
```

# Data Preparations

```python
x = df[['RM','LSTAT','PTRATIO']]
x.head()
```

```python
y = df['MEDV']
y.head()
```

### Test Train Split - Always do this before scaling the data to avoid leaking data. If you do it after, the scalar will have already seen the data.
After checking collinearity


```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

## Scale data - After test train split
Check the different types of scalers<br>
Scaler does- There are multiple columns, eg. x(distance (500 kms), fuel used (6o liters), state of road (bad or good)) and y(price)
Our data has different values. For them to be in the same scale, you have to use the same scale. So that the value of 500 does not affect the model cause it's big but it is not actually the most important value.

- MIN-MAX SCALER- when you know the range of the data eg 0-100. For example, for pixels colors are always from 0-255. Does not work for multiple columns with different data types. Mainly used for image processing. 
- Standard scaler- standardizes all the data and comes up with a scale. Most commonly used

You can mfeed without scaling but your model may not be as accurate.


```python
scaler = StandardScaler()
```

```python
# scaler.fit()
# scaler.transform()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
#IMPORTANT NOTES!!!!!!!!!
# train data with fit transform 
# test data with test data
# We only scale our predictors, we don't scale the predicted value
x_test_scaled = scaler.transform(x_test)
```

## Modeling

```python
model = LinearRegression()
```

```python
# training the model- TRAIN THE MODEL WITH ONLY THE TRAIN DATA
model.fit(x_train_scaled,y_train)
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
df = pd.DataFrame({"true":y_test,"pred":y_pred})
df.head()
```

## Validation 

validating models perfomance

```python
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
r2  = r2_score(y_test,y_pred) #RMSE

#MAE AND R2 SCORE ARE THE MOST INTERPRETABLE. MAE SHOWS HOW FAR YOUR PREDICTED 
#ERROR IS FROM YOUR ACTUAL ERROR
# mae 2.2802248550435116- PREDICTED VALUE IS 2.28 POINTS AWAY FROM THE ACTUAL VALUE
# R2 0.6548518476845391- 65% OF THE TIME, THE MODEL IS ACCURATE
```

```python
print("mse",mse)
print("mae",mae)
print("R2" ,r2)
```
